## **Investing Fandango Movie Ratings Project**

Uploading the datsets

```python
import pandas as pd
pd.options.display.max_columns=20

previous = pd.read_csv('/content/fandango_score_comparison.csv')

previous.iloc[4:10]
```

```python
after = pd.read_csv('/content/movie_ratings_16_17.csv')
after.iloc[2:8]
```

Extracting Fandango Informations

```python
fandango_previous=previous[['FILM', 'Fandango_Stars' , 'Fandango_Ratingvalue' , 'Fandango_votes' , 'Fandango_Difference']].copy()
fandango_previous.iloc[4:10]
```

```python
fandango_after=after[['movie' , 'year' , 'fandango']].copy()
fandango_after.iloc[2:8]
```

```python
fandango_previous['year']= fandango_previous['FILM'].str[-5:-1]
fandango_previous.iloc[4:10]
```

```python
fandango_previous['year'].value_counts()
```

```python
fandango_2015=fandango_previous[fandango_previous['year']=='2015']
fandango_2015['year'].value_counts()
```

```python
fandango_after['year'].value_counts()
```

```python
fandango_2016=fandango_after[fandango_after['year']==2016].copy()
fandango_2016['year'].value_counts()
```

Visualization

```python
import matplotlib.pyplot as plt
plt.style.use('tableau-colorblind10')

fandango_2015['Fandango_Stars'].plot.kde(label='2015', legend=True, figsize=(6,4))
fandango_2016['fandango'].plot.kde(label='2016', legend=True)

from numpy import arange

plt.title('Distribution of 2015 and 2016 Fandango ratings ')
plt.xlabel('Stars')
plt.xlim(0,5)
plt.xticks(arange(0,5.1,.5))
plt.ylabel('Density')
plt.ylim(0,1)
plt.yticks(arange(0,1.1,.2))
plt.show()
```

```python
print('2015 Ratings Distribution')
print(fandango_2015['Fandango_Stars'].value_counts(normalize=True).sort_index()*100)
```

```python
print('2016 Ratings Distribution')
print(fandango_2016['fandango'].value_counts(normalize=True).sort_index()*100)
```

```python
norm_2015 = fandango_2015['Fandango_Stars'].value_counts(normalize=True).sort_index()*100
norm_2016 = fandango_2016['fandango'].value_counts(normalize=True).sort_index()*100

df_freq = pd.DataFrame({'2015':norm_2015, '2016':norm_2016})

df_freq.plot.bar(figsize=(6,4))
plt.title('Frequency of 2015 and 2016 Fandango ratings')
plt.xlabel('Stars')
plt.ylabel('Frequency (%)')
plt.show()

```

```python
mean_2015 = fandango_2015['Fandango_Stars'].mean()
mean_2016 = fandango_2016['fandango'].mean()

median_2015 = fandango_2015['Fandango_Stars'].median()
median_2016 = fandango_2016['fandango'].median()

mode_2015 = fandango_2015['Fandango_Stars'].mode()[0]
mode_2016 = fandango_2016['fandango'].mode()[0]


summary = pd.DataFrame()

summary['2015'] = [mean_2015, median_2015, mode_2015]
summary['2016'] = [mean_2016, median_2016, mode_2016]
summary.index = ['mean', 'median', 'mode']
summary
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('tableau-colorblind10')

# Bar positions
x = np.arange(len(summary.index))
width = 0.35

plt.figure(figsize=(6,4))

# 2015 bars (shift left)
plt.bar(x - width/2, summary['2015'], width=width, label='2015')

# 2016 bars (shift right)
plt.bar(x + width/2, summary['2016'], width=width, label='2016')

plt.title('Comparing Summary Statistics of 2015 and 2016 Fandango Ratings')
plt.xlabel('Statistics')
plt.ylabel('Stars')
plt.xticks(x, summary.index)
plt.legend()

plt.show()
```
